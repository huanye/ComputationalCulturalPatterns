### Introduction

The common practice period in western classical music has been specified by musicologists as the period of tonal music. The period lasted for about 300 years consisting mainly of three eras in western music history: the Baroque era from 1600 to 1730, the Classical era from 1730 to 1830 and the Romantic era from 1830 to 1900 [[1]](https://en.wikipedia.org/wiki/Common_practice_period). This partition of the three musical eras in the common practice period represents a paradigm shift in music, along with the evolution of societies. 

In the Baroque era, composers believed that music could act as a potent tool of communication, which was one legacy from the ancient Greece and Rome, and the function of communication was signified by the widely application of polyphony as the musical texture in this era. However, this kind of communication was subject to the limitations imposed by the social context: composers in the Baroque era lived on the patronage of religious and political institutions which dictated what kind of music the composers could produce. Composers had no autonomy but only to meet the functional requirements by their patrons, and their roles were signified by the lack of autonomy in their compositions which strictly followed certain composing rules and musical forms. For example, the form of cantatas by J.S.Bach was established mainly to cater for liturgical demands of the Leipzig church that employed him, but not necessarily because Bach himself found the form intriguing [[2]](https://en.wikipedia.org/wiki/Cantata#Bach_cantatas). 

When the Classical Era came, composers became less constrained by their patrons thanks to the growing number of public music performance that most composers could make their ends meet. Music was no longer limited within courts and churches in the service of aristocrats and religon but expanded its circle to the growing middle class due to the gradual booming economy in Europe brought by the Industrial Revolution. More composers therefore worked as freelancers such as Mozart and Beethoven. Consequently their works were found to have fully reflected composers' musical ideas, which could be best signified by Beethoven's three compositional periods with increasing number of personal experiments of his own musical understanding. The growing autonomy of music could be evidenced by the musical forms in this period focusing more on the overall struture and clear melody line compared with the detailed ornamentations and counterpoints between various melody lines in the Baroque Era [[3]](https://en.wikipedia.org/wiki/Classical_period_(music)#Baroque/Classical_transition_c._1730%E2%80%931760). 

This trend of growing personal expression in music extended to the Romantic Era, when music performance increasingly appeared in salons and private parties between friends, and more composers played dual roles of both the music creator and performer. Recitals became more prevalent and performers prided themselves on their virtuosity playing various musical instruments through which personal emotions were fully articulated. Emotional expression was considered of the utmost importance in this era, which represented a rebel against social and political norms of the Age of Enlightenment and the scientific rationalization [[4]](https://en.wikipedia.org/wiki/Romantic_music). This kind of dramatic emotional expression was signified by being less restrained by musical forms and more emphasis on the "extra-musical narrative" in music, or the program music [[5]](https://en.wikipedia.org/wiki/Program_music).


### Culture patterns and sign relations in Peircean language

Music is traditionally considered an integral part of human culture. As introduced above, it can be the tools for communication of thoughts, self contemplation and rationalization, and expression of personal emotions, all of which contributes to the rediscovery of humanity and connection to humanity. The common practice of music composing as a culture phenomenon attained its prime time in Europe from the 17th to the 19th century, shaped people's hearing habits and aesthetic values of music, and still has the lasting influence on people's music perception today. Each musical era in the common practice period had its own social and historical context as presented in the introduction, and was thus characterized by certain music style one may perceive in music compositions from that era. So in this sense, the music styles of different eras could be the signs representing the cultural pattern of the three musical eras, which belongs to a conventionalized pattern because the pattern of three musical eras has been widely acknowledged and identified by musicology and relevant theory in aesthetics. 

For the sign relations using the Peirce's framework [6], I consider two layers of the triadic relations here. The first layer relates the music style to its social context. More specifically, the style of each musical era could be one sign vehicle of the general socioeconomic condition and cultural disposition of composers in that era, which is the object, and the interpretant is that the musical style reflects both the composer's socioeconomic status and his or her musical understanding. However, I need to further analyze and operationalize the musical style, as the style by itself could be an abstract object to describe and quantify. So I apply Peirce's distincion between 'type' and 'token' here, and construct some metrics based on musical elements such as rhythm, pitch,texture and tonality as the "tokens", and the music style as the "type" to build the second layer of the triadic relation: the tokens, or the constructed metrics, are signs of the "type", or the music style, which is the object, and the interpretant is that the metrics derived from a composition can be the defining featues charactrizing the music style of the composition. So the next step is to find out relevant features which can capture and distinguish the music style changes in the common practice period.

### Interpretation of features and music style predictions 

First of all, one apparent trend can be observed from the sequence of three musical eras as the introduction presents: musical compositions became more "autonomous", which was evidenced in two aspects: 

a) Musical "rules" and rhethoric became more flexible with more diversified musical forms. 

b) Growing demands for emotional expressions in music called for more dramatic changes and higher density of various musical elements.  

Having these two guidelines in hand, I construct the following 6 metrics to characterize the music style of a compositions:

1) the "normalized Pairwise Variability Index", or NPIV which was initially applied to speech recognition [7], and the index is defined as below:

$$nPIV = \frac{100}{m-1}\sum_{k=1}^{m-1}|\frac{d_k-d_{k-1}}{d_k+d_{k-1}}|$$

where m is number of notes in one part of the music composition and $d_k$ is the duration of the kth note. We can see from the formula that the index is in essence an aggregation of local changes in note duration by taking the average of all the neighboring differences in note duration. 

This metric mainly describe the rhythmic change in a composition, which is related to both evidences a) and b). Because more frequent rhythmic changes could be an indication of a more flexible musical form, and also contribute to the dramatic emotional expression.

2) the chord proportion in a composition, which is defined as the the fraction of chords in total number of notes from a composition. 

This feature mainly indicated the texture of the music. As the chord is defined as a group of more than 3 notes sounded together, we expect it would appear less in the Baroque music which takes the polyphony as its main texture compared with the Classical and Romantic music. This feature is also related to both evidences because application of chord is one of the important music rhethoric, and also increase the note density for dramatic emotional expression.

3) the mean pitch difference in various parts of a composition, which is defined as the distance between the highest mean pitch and the lowest mean pitch among all parts of a composition. 

This feature is one of the metrics based on the pitches of the music. We expect the mean pitch difference would increase as the music style evolved in the common practice period. This feature is mainly related to evidence b) because the higher pitch difference among all parts, the more easily the music can build the tension and express strong emotions. 


4) the number of major pitch classes, where the major pitch classes are defined as those appearing more than 9% of the total number of notes.

This feature is another metric based on the pitches of the music. We expect the number of major pitch classes would decrease as the music style evolved in the common practice period because the increase in diversity of pitch classes from the Baroque Era to the Romantic Era would lead to a decrease in the number of major pitch classes. This feature is related to both evidence a) and b) because the number of major pitch classes is an indication of pitch class concentration, which is related to the flexibility of musical form, and also affect the dramatic emotional expression as high pitch class concentration could hinder the radical jump and sharp constrast in the articulation. 

5) The pitch range, which literally means the distance between the highest pitch and the lowest pitch in a composition.

This feature is another metric based on the pitches of the music. We expect the pitch range would expand as the music style evolved in the common practice period. One reason is the technology progress leading to the further innovation of the musical instrument to allow larger pitch range for improving the expressiveness of the instrument. This feature is mainly related to evidence b) because a larger pitch range literally reflects the more dramatic change in pitches to incorporate higher level of tension and contrast in a composition for emotional expression.

6) The fraction of perfect fifth interval in a composition, which is defined as the the fraction of the perfect fifth intervals in 9% of the total number of notes in a compositions.

This feature is a metric describing the tonality of a composition. The most salient property defining the tonality of music in the common practice period is the perfect fifth inverval, where the first dominant note representing the tension building and the second tonic note representing the tension releasing. We expect the fraction of perfect fifth interval would increase as the music style evolved in the common practice period because the more perfect fifth intervals indicate a more frequent modulation and thus the less stable tonality of music, which became more apparent as time went by. This feature is also related to both the evidence a) and b) because the frequent modulation indicated by the large number of perfect fifth intervals implies both a more flexible music form and also contributes to the dramatic emotional expression signified by the less stable tonality in the composition.

Next, I implement all the six features one by one and apply them to the training data to check their discerning power of the three music styles in the common practice period. First I define a function taking the MusicXML file as the input and output a tuple of four lists for further metric calculations.


```python
import numpy as np
import pandas as pd
import xml.etree.ElementTree as et
import matplotlib.pyplot as plt
import seaborn as sns
import os

# the dictionary translate various note durations into numeric values 
# for relevant metric calculations. 

pitch_weights = {"breve":2,"whole":1,"half":0.5,"quarter":0.25,
                "eighth":0.125,"16th":0.0625,"32nd":0.03125,
                "64th":0.015625,"128th":0.0078125}

def getPitchDuration(filename,pitch_weights):
    '''
    The function takes the MusicXML file as the input and output a tuple of 
    four lists for further metric calculations:
    1) the list of all note pitches from the part with the highest mean pitch.
    2) the list of all note durations from the part with the highest mean pitch.
    3) the list of all pitch means from all parts. 
    4) the list of booleans indicating if a true chord exists for each chord event.  
    '''
    xtree = et.parse(filename)
    xroot = xtree.getroot()
    dweights_ = None
    pitches_ = None
    highest_pitches = 0
    score = xroot.find("Score")
    pitch_means = []
    is_chord = []
    for part in score.findall("Staff"):
        pitches=[]
        dweights = []
        tie_stack = []
        for measure in part.findall("Measure"):
            for chord in measure.findall("Chord"):
                if len(chord.findall('Note'))>1:
                    is_chord.append(1)
                else:
                    is_chord.append(0)
                dweight = pitch_weights[chord.find("durationType").text]
                if chord.find("dots")!=None:
                    dweight = dweight*1.5

                lastnote = chord.findall('Note')[-1]
                if lastnote.find("endSpanner")!=None:
                    while len(tie_stack)>0:
                        weight = tie_stack.pop()
                        dweight+=weight
                    dweights.append(dweight)
                    pitches.append(int(lastnote.find('pitch').text))
                    
                elif lastnote.find("Tie")!=None:
                    tie_stack.append(dweight)
                else:
                    pitches.append(int(lastnote.find('pitch').text))
                    dweights.append(dweight) 
        pitch_means.append(np.mean(pitches))
        if np.mean(pitches)>highest_pitches:
            highest_pitches = np.mean(pitches)
            pitches_ = pitches
            dweights_ = dweights
        
    
    return pitches_,dweights_,pitch_means,is_chord
```

Then I define several metric functions charactering the music styles in the common practice period as explained above. 

```python
def pitchClass(pitch):
    '''
    This function translates the pitch value from the midi file to 
    the position of the that pitch in the circle of fifths[8].
    '''
    class_list = ["C","G","D","A","E","B","F#","C#","G#","D#","A#","F"]
    order = sorted(class_list)[3:]+sorted(class_list)[:3]
    return class_list.index(order[pitch%12])

def meanDiff(pitch_means):
    '''
    This function computes a metric of pitch difference in all parts of
    a composition.
    '''
    return max(pitch_means)-min(pitch_means)

def chordFrac(is_chord):
    '''
    The function returns the metric of the fraction of chords 
    in total number of notes from a composition.
    '''
    return np.mean(is_chord)

def domiPitchFraction(pitches):
    '''
    The function returns the metric of the fraction of the dominant pitch 
    classe in total number of notes from a composition.
    '''
    import scipy.stats as stats
    pclasses = np.array(list(map(pitchClass,pitches)))
    frac_domi = sum(pclasses==stats.mode(pclasses).mode[0])/len(pclasses)
    return frac_domi

def numMostPclass(pitches):
    '''
    The function returns of the number of the major pitch classes in a
    composition, where the major pitch classes are those which appear 
    more than 9% of the total number of notes.
    '''
    pclasses = pd.Series(list(map(pitchClass,pitches)))
    return sum(pclasses.value_counts()>len(pitches)*0.09)


def pitchRange(pitches):
    '''
    The function returns the distance between the highest pitch and 
    the lowest pitch in a composition.
    '''
    return max(pitches) - min(pitches)

def fracPerfectFifth(pitches):
    '''
    The function returns the fraction of the perfect fifth intervals in 
    9% of the total number of notes in a compositions.
    '''
    pclasses = pd.Series(list(map(pitchClass,pitches)))
    diff = np.diff(pclasses)
    num_fifth = sum((diff==1) | (diff==-11))
    return (num_fifth)/(len(pitches)*0.09)


def computeLeap(pitches):
    '''
    The function returns the metric of the average leap for a part 
    of a composition, where the leap is the distance between neighboring 
    pitches. 
    '''
    leaps = []
    for i in range(1,len(pitches)-1):
        diff1 = abs(pitches[i]-pitches[i-1])
        diff2 = abs(pitches[i]-pitches[i+1])
        leaps.append(min(diff1,diff2))
    if len(leaps)==0:
        return None
    return np.mean(leaps)

def computeNPIV(weights):
    '''
    The function computes the metric of average difference in neighboring
    note durations for a part of a composition.
    '''
    m  = len(weights)
    weights = np.array(weights)/weights[0]
    abs_sum=0
    if m-1<1:
        return None
    for i in range(m-1):
        abs_sum+=(abs(weights[i]-weights[i+1])/(weights[i]+weights[i+1]))
    
    return 100*abs_sum/(m-1)

def getStatsEra(directoryName,func,index=0):
    '''
    The function takes a directory of MusicXML files and a specific metric 
    function as the input, and returns a list of the metric values for all
    files in the directory.
    '''
    results = []
    for filename in os.listdir(directoryName):
        if filename.endswith(".mscx"):
            result = func(getPitchDuration(directoryName+filename,pitch_weights)[index])
            if result!=None:
                results.append(result)
    return results
```

Next I apply the six metrics one by one to the compositions from the three eras, and check the discerning power of each metric by plotting the metric histograms for each musical era. First we check the fraction of chords in a composition.

```python
baroque_chord = np.random.choice(getStatsEra("../baroque/",chordFrac,3),77).tolist()
classical_chord = getStatsEra("../classical/",chordFrac,3)
romantic_chord = np.random.choice(getStatsEra("../romantic/",chordFrac,3),77).tolist()
kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})
sns.distplot(baroque_chord, color="skyblue", label="baroque", **kwargs)
sns.distplot(classical_chord, color="orange", label="classical", **kwargs)
sns.distplot(romantic_chord, color="deeppink", label="romantic", **kwargs)
plt.legend()
plt.show()
```

We can see the result is what we expect: the fractions of chords would increase as the music style evolved in the common practice period. Next we check the metric of the mean pitch difference in various parts of a composition.

```python
baroque_pdiff = np.random.choice(getStatsEra("../baroque/",meanDiff,2),77).tolist()
classical_pdiff = getStatsEra("../classical/",meanDiff,2)
romantic_pdiff = np.random.choice(getStatsEra("../romantic/",meanDiff,2),77).tolist()
kwargs = dict(hist_kws={'alpha':.6}, kde_kws={'linewidth':2})
sns.distplot(baroque_pdiff, color="skyblue", label="baroque", **kwargs)
sns.distplot(classical_pdiff, color="orange", label="classical", **kwargs)
sns.distplot(romantic_pdiff, color="deeppink", label="romantic", **kwargs)
plt.legend()
plt.show()
```

It turns out that the result is not what we expected in that the mean pitch distance of the Classical music is smaller that that of the Baroque music. The reason may be the sampling bias in my training data. Next we check the metric of the nPIV.

We can see the trend roughly matches what we expected although the overlaps between different musical eras are apparent. Next we check the metric of the number of major pitch classes.

However, for prediction of music styles in the common practice period, those features without discerning power should also be considered because they may still be good predictors when combined together to predict the music styles. 

So next I use all 6 features in my prediction model. First I load and compute the test data for each feature in each musical era.    

```python
baroque_npiv_test = getStatsEra("../piano/baroque/",computeNPIV,1)
classical_npiv_test = getStatsEra("../piano/classic/",computeNPIV,1)
romantic_npiv_test = getStatsEra("../piano/romantic/",computeNPIV,1)

baroque_npitch_test = getStatsEra("../piano/baroque/",numMostPclass)
classical_npitch_test = getStatsEra("../piano/classic/",numMostPclass)
romantic_npitch_test = getStatsEra("../piano/romantic/",numMostPclass)

baroque_pdiff_test = getStatsEra("../piano/baroque/",meanDiff,2)
classical_pdiff_test = getStatsEra("../piano/classic/",meanDiff,2)
romantic_pdiff_test = getStatsEra("../piano/romantic/",meanDiff,2)

baroque_chord_test = getStatsEra("../piano/baroque/",chordFrac,3)
classical_chord_test = getStatsEra("../piano/classic/",chordFrac,3)
romantic_chord_test = getStatsEra("../piano/romantic/",chordFrac,3)

baroque_prange_test = getStatsEra("../piano/baroque/",pitchRange)
classical_prange_test = getStatsEra("../piano/classic/",pitchRange)
romantic_prange_test = getStatsEra("../piano/romantic/",pitchRange)

baroque_pfifth_test = getStatsEra("../piano/baroque/",fracPerfectFifth)
classical_pfifth_test = getStatsEra("../piano/classic/",fracPerfectFifth)
romantic_pfifth_test = getStatsEra("../piano/romantic/",fracPerfectFifth)

```

Next I put the training data and test data into two dataframes.

```python
len_baroque = len(baroque_npiv)
len_classical = len(classical_npiv)
len_romantic = len(romantic_npiv)

df_train = pd.DataFrame({"npiv":baroque_npiv+classical_npiv+romantic_npiv,
                          "npitch":baroque_npitch+classical_npitch+romantic_npitch,
                          "pdiff":baroque_pdiff+classical_pdiff+romantic_pdiff,
                          "chord":baroque_chord+classical_chord+romantic_chord,
                          "prange":baroque_prange+classical_prange+romantic_prange,
                          "pfifth":baroque_pfifth+classical_pfifth+romantic_pfifth})


len_baroque_test = len(baroque_npiv_test)
len_classical_test = len(classical_npiv_test)
len_romantic_test = len(romantic_npiv_test)

df_test = pd.DataFrame({"npiv":baroque_npiv_test+classical_npiv_test+romantic_npiv_test,
                         "npitch":baroque_npitch_test+classical_npitch_test+romantic_npitch_test,
                          "pdiff":baroque_pdiff_test+classical_pdiff_test+romantic_pdiff_test,
                          "chord":baroque_chord_test+classical_chord_test+romantic_chord_test,
                          "prange":baroque_prange_test+classical_prange_test+romantic_prange_test,
                          "pfifth":baroque_pfifth_test+classical_pfifth_test+romantic_pfifth_test})


```

Then apply the logistic regression model to the data using the sklearn library for prediction, and check the accuracy of the prediction.

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegressionCV
```

```python
X_train = df_train.to_numpy()
y_train = [0]*len_baroque+[1]*len_classical+[2]*len_romantic
X_test = df_test.to_numpy()
y_test = [0]*len_baroque_test+[1]*len_classical_test+[2]*len_romantic_test
```

```python
clf = LogisticRegressionCV(cv=10, random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

The accuracy of 0.76 for a ternary classification is not bad, but also not perfect. Next we check if all six features are statistically sigificant: 

```python
from sklearn.feature_selection import chi2
scores, pvalues = chi2(X_train, y_train)
print(pvalues)
```

```python
def composer2D(func1,func2,fname1,fname2,composers,index=0):
    '''
    The function scatterplots the medians of all values using two metrics 
    specified as the input func1 and func2 for each composer.
    '''
    median1s = []
    median2s = []
    schools = []
    names = []
    for composer in composers:
        names.append(composer)
        lst1 = getStatsEra("../piano/"+composer+"_small/",func1,1)
        median1s.append(np.median(lst1))
        lst2 = getStatsEra("../piano/"+composer+"_small/",func2,index)
        median2s.append(np.median(lst2))
        schools.append(composers[composer])
    df = pd.DataFrame({fname1:median1s,fname2:median2s,"Era":schools,"name":names})
    
    plt.figure(figsize=(20,5), dpi= 80)
    p1 = sns.scatterplot(data=df, x=fname1, y=fname2, hue="Era",s=100)
    for line in range(0,df.shape[0]):
        p1.text(df[fname1][line]+0.2, df[fname2][line], df["name"][line], horizontalalignment='left', size='medium', color='black')
    plt.show()
```

We can find 3 common patterns in all the graphs above: 1)composers from the Baroque Era and composers from the Classical Era cannot be easily differentiated easily based on their locations in the graphs. 2) However, composers from the Romantic Era and the Impressionist Era can be clearly separated from composers from the previous two musical eras. 3) Beethoven's position in almost all graphs is located on the boundary line between all composers after his living years and all composers before his living year, which makes me believe Beethoven can be a good "watershed" for the binary classification as to whether a composition was composed before or after Beethoven's living year.
```python
# compute feature values for test data
before_npiv_test = getStatsEra("../piano/before/",computeNPIV,1)
after_npiv_test = getStatsEra("../piano/after/",computeNPIV,1)
before_dpitch_test = getStatsEra("../piano/before/",domiPitchFraction)
after_dpitch_test = getStatsEra("../piano/after/",domiPitchFraction)
before_npitch_test = getStatsEra("../piano/before/",numMostPclass)
after_npitch_test = getStatsEra("../piano/after/",numMostPclass)
before_prange_test = getStatsEra("../piano/before/",pitchRange)
after_prange_test = getStatsEra("../piano/after/",pitchRange)
before_pfifth_test = getStatsEra("../piano/before/",fracPerfectFifth)
after_pfifth_test = getStatsEra("../piano/after/",fracPerfectFifth)
before_pdiff_test = getStatsEra("../piano/before/",meanDiff,2)
after_pdiff_test = getStatsEra("../piano/after/",meanDiff,2)
before_chord_test = getStatsEra("../piano/before/",chordFrac,3)
after_chord_test = getStatsEra("../piano/after/",chordFrac,3)
```

Then we again use the logistic regression model to make the binary prediction

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.linear_model import LogisticRegressionCV
```

```python
df_train_new = pd.DataFrame({"npiv":before_npiv+after_npiv,
                          "npitch":before_npitch+after_npitch,
                          "prange":before_prange+after_prange,
                          "pfifth":before_pfifth+after_pfifth,
                            "pdiff":before_pdiff+after_pdiff,
                            "chord":before_chord+after_chord})
df_test_new = pd.DataFrame({"npiv":before_npiv_test+after_npiv_test,
                          "npitch":before_npitch_test+after_npitch_test,
                           "prange":before_prange_test+after_prange_test,
                          "pfifth":before_pfifth_test+after_pfifth_test,
                           "pdiff":before_pdiff_test+after_pdiff_test,
                            "chord":before_chord_test+after_chord_test})

len_before = len(before_npiv)
len_after = len(after_npiv)
len_before_test = len(before_npiv_test)
len_after_test = len(after_npiv_test)


X_train = df_train_new.to_numpy()
X_test = df_test_new.to_numpy()
y_train = [0]*len_before+[1]*len_after
y_test = [0]*len_before_test+[1]*len_after_test


clf = LogisticRegressionCV(cv=10, random_state=0).fit(X_train, y_train)
y_pred = clf.predict(X_test)
#check the confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

For now, all features I choose for music style identification and prediction are derived from the midi files which I collected as the raw data, and this kind of data are symbolic data [9], and these feature are consequently the symbolic features derived from the symbolic data to interpret music styles in the common practice period. An alternaive group of signs for interpretation of music styles can be derive from the audio data which directly represents sounds using encoded digital samples from the analog sound waves. Examples of audio features include amplitude envelopes, energy, timbral texture and temporal features. Because these feature are related to the physical property of the sound, they are referred to as the "low-level" representations. In contrast, symbolic features are based on the musical events and paramters such as rhythm, pitch, melody and tonality, so they are referred to as the "high-level" representationa. The main reason I use these "high-level" features for the interpretation of music styles is that they have explicit corresponding musicological meanings, and can be more easily related to the particular period in the western music history, and thus can be applied for both the theoretical exploration and more practical applications.

### Conlusion

Music styles in the common practice period has been investigated by musicologiets for years. A long list of features has been proposed to characterize each musical era [9]. This project takes an empirical approach to identify the cultural pattern of music styles signified by a much smaller feature space derived from basic music elements using symbolic data. The layer of sign relations between the symbolic features, music styles and their interpretant begets the next layer of sign relations between music styles, the socioeconomic status of composers and their interpretant, which builds the bridge linking the symbolic features to their social contexts. Accordingly, musical functions symbolically constructed by those musical features could be attached to their humanistic values and social roles. For example, the Baroque music was a tool for communication and religious or political aggregation. With music, people became more disciplined and easily controlled by their superiors, so music played its role for those superiors in the execution and maintenance of their power. The Classical music could be a medium for self contemplation and intellectual inspiration. With music, people rebuilt their own identity, cultivated their thinking ability and   rationalization, refused to blindly obey the rules from their superiors, so music then became the catalyst for the Age of Enlightenment.  The Romantic music was regarded as the expressive outlet for personal emotion. With music, people expressed their ardent affection, desperate sorrow or strong anger, so music became the rhetorical vehicle for people's voice to be heard and constributed to the respect for personal rights in the society. To sum up, the music and the society are intertwined, as manifested by Simon Frith's proposition for the function of music: the first function is self definition and of creating a place in society for people, and the second function is "managing the relationship between our public and private emotional life"[10][11]
